```python
df2 =pd.read_csv('hiring.csv')
df2.dtypes
```

```python
from word2number import w2n
```

```python
def string_to_number(input_string):
    try:
        str(input_string)
        num_value = w2n.word_to_num(input_string)
        return num_value
    except ValueError:
        return None
```

```python
df2['experience'] = df2['experience'].apply(string_to_number)
```

```python
df2
```

```python
df2.fillna((df2.mean()), inplace=True)
```

```python
df2['experience']=df2['experience'].apply(round)
```

```python
df2
```

```python
df2.columns
```

```python
experience = df2['experience']
testscore = df2['test_score(out of 10)']
score =df2['interview_score(out of 10)']
salary =df2['salary($)']
```

```python
reg=linear_model.LinearRegression()
reg.fit([experience,testscore,score],salary)
```
